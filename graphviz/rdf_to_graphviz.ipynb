```python
import io
import graphviz
import pydotplus
from IPython.display import display, Image
from rdflib import Graph, URIRef, Literal
from rdflib import Graph, tools
from rdflib.tools import rdf2dot, rdfs2dot
from typing import List, Dict, Set, Tuple
import networkx as nx
from utils import *
import random
```

```python
g = Graph()
g.parse((DATA_DIR / "pred_path.ttl").resolve())
```

```python
def visualize_v1(g: Graph) -> None:
    stream = io.StringIO()
    tools.rdf2dot.rdf2dot(g, stream, opts = {display})
    dg = pydotplus.graph_from_dot_data(stream.getvalue())
    dg.set_size('"15,8!"')

    # displaying image seemed to sometimes run into memory errors, so just save the image as a
    # png then have jupyter display it separately
    png = dg.create_png()
    display(Image(png))
```

```python
def visualize_v2(g: Graph) -> None:
    stream = io.StringIO()
    tools.rdf2dot.rdf2dot(g, stream, opts = {display})
    dg = pydotplus.graph_from_dot_data(stream.getvalue())
    dg.set_size('"15,8!"')

    # displaying image seemed to sometimes run into memory errors, so just save the image as a
    # png then have jupyter display it separately
    png = dg.create_png()
    display(Image(png))
```
